### Revisión y calidad de datos

**🎯 Objetivo:** Detectar y corregir problemas en los datos que puedan afectar el análisis de revenue, costos y rentabilidad.

Se revisan los 3 datasets

```python
catalog = catalog.rename(columns={'costo_unitario': 'precio_unitario'})

print()
catalog.info()
catalog.head(10)
```

```python
orders['fecha_hora_pedido'] = pd.to_datetime(orders['fecha_hora_pedido'], errors='coerce')
orders['cantidad'] = orders['cantidad'].astype('Int64')
orders['pais'] = orders['pais'].str.title()
orders = orders.rename(columns={'fecha_hora_pedido': 'fecha_pedido'})
orders = orders.drop_duplicates()

precios_correctos = catalog[['nombre_producto', 'precio_unitario']].rename(
    columns={'precio_unitario': 'precio_correcto'}
)
orders = orders.merge(precios_correctos, on='nombre_producto', how='left')
orders['precio_unitario'] = orders['precio_correcto']
orders = orders.drop('precio_correcto', axis=1)

orders['cantidad'] = orders['cantidad'].fillna(1)
orders['monto_descuento'] = orders['monto_descuento'].fillna(0)

columnas_categoricas = ['pais', 'dispositivo', 'fuente_referencia', 'nombre_producto', 'categoria_producto']

for col in columnas_categoricas:
    orders[col] = orders[col].fillna('No especificado')

categorias_correctas = catalog[['nombre_producto', 'categoria_producto']]
orders = orders.merge(categorias_correctas, on='nombre_producto', how='left', suffixes=('', '_correcta'))

mask_no_especificado = orders['categoria_producto'] == 'No especificado'
orders.loc[mask_no_especificado, 'categoria_producto'] = orders.loc[mask_no_especificado, 'categoria_producto_correcta']

orders = orders.drop('categoria_producto_correcta', axis=1)

orders['monto_total'] = (
    orders['cantidad'] * orders['precio_unitario'] - 
    orders['monto_descuento']
)

orders = orders[~((orders['nombre_producto'] == 'No especificado') & (orders['precio_unitario'].isnull()))]

orders['categoria_producto'] = orders['categoria_producto'].replace({
    'Electronica': 'Electrónica'
})  

print()
orders.info()
orders.head(10)
```

```python
marketing['fecha'] = pd.to_datetime(marketing['fecha'], errors='coerce')
marketing['pais'] = marketing['pais'].str.title()
marketing = marketing.drop_duplicates()

def extraer_canal(id_campaña):
    if pd.isna(id_campaña):
        return 'No especificado'
    elif id_campaña.startswith('organic_'):
        return 'organic'
    elif id_campaña.startswith('paid_search_'):
        return 'paid_search'
    elif id_campaña.startswith('social_'):
        return 'social'
    else:
        return 'No especificado'

mask_canal_faltante = marketing['canal'].isna()
marketing.loc[mask_canal_faltante, 'canal'] = marketing.loc[mask_canal_faltante, 'id_campaña'].apply(extraer_canal)

marketing.info()
marketing.head(10)
```

## 🔹 Paso 2: Analizar si el negocio es rentable

### 2.1 Cálculo de KPIs principales

**🎯 Objetivo:** Calcular los indicadores clave del negocio para evaluar ingresos, costos y rentabilidad.

Se usan los 3 datasets (`orders`, `catalog`, `marketing_spend`):

**📊 Parte 1: Rentabilidad del negocio**
- ¿Cuál es el ingreso total (revenue)? 
- ¿Cuál es el costo total? 
- ¿Cuánto se ha invertido en marketing? 
- ¿El negocio es rentable? (calcular profit)  

---

**📈 Parte 2: Comportamiento de ventas**
- ¿Cuál es el ticket promedio por orden? 
- ¿Cuál es la cantidad promedio de productos por orden? 
- ¿Cuál es el producto más vendido?
- ¿Cuánto se ha gastado en marketing por canal? 

```python
print("RENTABILIDAD DEL NEGOCIO")

print("\nIngreso Total (Revenue)")
revenue = orders['monto_total'].sum()
print(f"${revenue:,.2f}")

print("\nCostos Totales")
costos = orders['cantidad'] * orders['precio_unitario']
costo_total = costos.sum()
print(f"${costo_total:,.2f}")

print("\nInversión de marketing total")
inversion_marketing = marketing['gasto'].sum()
print(f"${inversion_marketing:,.2f}")

print("\n¿El negocio es rentable?")
profit = revenue - costo_total - inversion_marketing
print(f"Profit = Revenue - Costos - Marketing")
print(f"Profit = ${profit:,.2f}")

if profit > 0:
    print("✅ El negocio ES rentable")
else:
    print("❌ El negocio NO es rentable")
```

```python
print("COMPORTAMIENTO DE VENTAS")

print("\nTicket Promedio")
ticket_mean = orders['monto_total'].mean()
print(f"${ticket_mean:,.2f}")

print("\nProductos promedio por orden")
order_mean = orders['cantidad'].mean()
print(f"{order_mean:,.2f}")

print("\nProducto más Vendido")
product_mas = orders.groupby('nombre_producto')['cantidad'].sum().sort_values(ascending=False)
print(product_mas)

print("\nGastos totales por canal de Marketing")
gasto_canal = marketing.groupby('canal')['gasto'].sum().sort_values(ascending=False)
for canal, gasto in gasto_canal.items():
    print(f"{canal}: ${gasto:,.2f}")
```

```python
# PARTE 1: Totales del funnel
# ======================

query_totals = '''
WITH cte_first AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'first_visit'
),
cte_item AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'select_item'
),
cte_cart AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'add_to_cart'
),
cte_check AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'begin_checkout'
),
cte_payment_info AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'add_payment_info'
),
cte_purch AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'purchase'
)
SELECT
  (SELECT COUNT(*) FROM cte_first) AS first_visit_users,
  (SELECT COUNT(*) FROM cte_item) AS select_item_users,
  (SELECT COUNT(*) FROM cte_cart) AS cart_users,
  (SELECT COUNT(*) FROM cte_check) AS checkout_users,
  (SELECT COUNT(*) FROM cte_payment_info) AS payment_info_users,
  (SELECT COUNT(*) FROM cte_purch) AS purchase_users;
'''

totals = pd.read_sql(query_totals, con=engine)
totals
```

```python
# PARTE 2: Conversiones
# ======================

query_conversion = '''

WITH cte_first AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'first_visit'
),
cte_item AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'select_item'
),
cte_cart AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'add_to_cart'
),
cte_check AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'begin_checkout'
),
cte_payment_info AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'add_payment_info'
),
cte_purch AS (
  SELECT DISTINCT id_usuario
  FROM events
  WHERE nombre_evento = 'purchase'
)
SELECT
  (SELECT COUNT(*) FROM cte_first) AS first_visit_users,
  (SELECT COUNT(*) FROM cte_item) AS select_item_users,
  (SELECT COUNT(*) FROM cte_cart) AS cart_users,
  (SELECT COUNT(*) FROM cte_check) AS checkout_users,
  (SELECT COUNT(*) FROM cte_payment_info) AS payment_info_users,
  (SELECT COUNT(*) FROM cte_purch) AS purchase_users,
  
((SELECT COUNT(*) FROM cte_first) - (SELECT COUNT(*) FROM cte_item)) * 100 
    / NULLIF((SELECT COUNT(*) FROM cte_first), 0) AS dropoff_first_visit_pct,
    
    ((SELECT COUNT(*) FROM cte_item) - (SELECT COUNT(*) FROM cte_cart)) * 100 
    / NULLIF((SELECT COUNT(*) FROM cte_item), 0) AS dropoff_item_pct,
       
    ((SELECT COUNT(*) FROM cte_cart) - (SELECT COUNT(*) FROM cte_check)) * 100 
    / NULLIF((SELECT COUNT(*) FROM cte_cart), 0) AS dropoff_cart_pct,
    
    ((SELECT COUNT(*) FROM cte_check) - (SELECT COUNT(*) FROM cte_payment_info)) * 100 
    / NULLIF((SELECT COUNT(*) FROM cte_check), 0) AS dropoff_check_pct,
    
    ((SELECT COUNT(*) FROM cte_payment_info) - (SELECT COUNT(*) FROM cte_purch)) * 100 
    / NULLIF((SELECT COUNT(*) FROM cte_payment_info), 0) AS dropoff_payment_pct;

'''

conversion = pd.read_sql(query_conversion, con=engine)
conversion
```

## 🔹 Paso 4: Evaluar si los usuarios regresan (retención por cohortes)

**🎯 Objetivo:** Analizar la retención de usuarios para entender si regresan después de registrarse.

**Tablas**

- `users` 
- `user_activity` 

---
1. Se identifica la cohorte de cada usuario según el **mes de registro**.


2. Se calcula la retención semanal: cuántos usuarios **se mantienen activos** en cada semana desde su registro.
   - `retenido_w1`: usuarios activos en la semana 1  
   - `retenido_w2`: usuarios activos en la semana 2  
   - `retenido_w3`: usuarios activos en la semana 3  


3. Se calcula el porcentaje de retención para cada semana, dividiendo los usuarios retenidos entre los clientes iniciales de la cohorte:  
   - `semana_1`: porcentaje de usuarios retenidos en la semana 1  
   - `semana_2`: porcentaje de usuarios retenidos en la semana 2  
   - `semana_3`: porcentaje de usuarios retenidos en la semana 3  

Se revisa que la columna de fecha esté en formato correcto (`DATE`).  
Se realiza la conversión usando: `CAST(fecha_registro AS DATE)`

```python
# Explorar tabla users
# =========================
query_users = '''
SELECT 
    id_usuario,
    CAST(fecha_registro AS DATE) as fecha_registro,
    país,
    dispositivo,
    tipo_plan
FROM users;
'''
users = pd.read_sql(query_users, con=engine)
users.info()
users.head(5)
```

```python
# Explorar tabla user_activity
# =========================
query_user_activity = '''
SELECT 
    id_usuario,
    CAST(fecha_actividad AS DATE) as fecha_actividad,
    dias_despues_registro,
    activo
FROM user_activity;
'''
user_activity = pd.read_sql(query_user_activity, con=engine)
user_activity.info()
user_activity.head(5)
```

```python

# Retención por cohortes
# ======================

query_cohort_retention_final = '''
WITH cohortes AS (
    SELECT
        u.id_usuario,
        DATE_TRUNC('week', CAST(u.fecha_registro AS DATE)) AS cohorte_semana,
        ua.dias_despues_registro,
        ua.activo
    FROM users u
    JOIN user_activity ua ON u.id_usuario = ua.id_usuario
),

usuarios_por_cohorte AS (
    SELECT 
        cohorte_semana,
        COUNT(DISTINCT id_usuario) AS clientes_iniciales
    FROM cohortes
    GROUP BY cohorte_semana
),

retencion AS (
    SELECT
        cohorte_semana, 
        COUNT(CASE WHEN dias_despues_registro BETWEEN 1 AND 7 AND activo = 1 THEN id_usuario END) AS retained_s1,
        COUNT(CASE WHEN dias_despues_registro BETWEEN 8 AND 14 AND activo = 1 THEN id_usuario END) AS retained_s2,
        COUNT(CASE WHEN dias_despues_registro BETWEEN 15 AND 21 AND activo = 1 THEN id_usuario END) AS retained_s3
    FROM cohortes
    GROUP BY cohorte_semana
)

SELECT
    TO_CHAR(r.cohorte_semana, 'YYYY-MM-DD') AS cohorte,
    u.clientes_iniciales,
    ROUND(r.retained_s1::numeric / u.clientes_iniciales, 2) AS semana_1,
    ROUND(r.retained_s2::numeric / u.clientes_iniciales, 2) AS semana_2,
    ROUND(r.retained_s3::numeric / u.clientes_iniciales, 2) AS semana_3
FROM retencion r
JOIN usuarios_por_cohorte u ON r.cohorte_semana = u.cohorte_semana
ORDER BY r.cohorte_semana;
'''

# Ejecutar la consulta
cohorte_final = pd.read_sql(query_cohort_retention_final, con=engine)
cohorte_final
```

## 🔹 Paso 5: Validar si los cambios generan impacto (test estadístico)

🎯 **Objetivo:** Evaluar si la modificación en la UI del checkout impacta la **tasa de conversión de compra**.

---

1. **Analizar el dataset** `experiment_checkout_ui.csv` para identificar la métrica principal **conversion**.
   - La métrica **conversion** es 1 si el usuario completó la compra, 0 si no.    
2. **Plantear la hipótesis estadística**     
3. **Aplicar el test estadístico adecuado** 
4. **Interpretar el resultado**  

---
Hipótesis estadística
   - **H₀ (Hipótesis nula):** La `UI Experimental` no presenta un impacto en las conversiones finales en comparación con la`UI Actual`.
   - **H₁ (Hipótesis alternativa):** La `UI Experimental` presenta un impacto diferente (mayor o menor) en las conversiones finales comparada con la `UI Actual`.
   
**Test estadístico:** 
Estadístico t: 0.81322, Valor P: 0.41611

**Nivel de significancia alpha:** 0.05

```python
# tu código aquí
import pandas as pd
from scipy.stats import ttest_ind
chkout = pd.read_csv('https://practicum-content.s3.amazonaws.com/datasets/experiment_checkout_ui.csv')
chkout.info()
chkout.head()
```

```python
tratamiento1 = chkout[(chkout['variante'] == 'tratamiento') & (chkout['convirtio'] == 1)]
tratamiento0 = chkout[(chkout['variante'] == 'tratamiento') & (chkout['convirtio'] == 0)]
control1 = chkout[(chkout['variante'] == 'control') & (chkout['convirtio'] == 1)]
control0 = chkout[(chkout['variante'] == 'control') & (chkout['convirtio'] == 0)]

print(f"Tratamiento - Convirtieron: {len(tratamiento1)}")
print(f"Tratamiento - No convirtieron: {len(tratamiento0)}")
print(f"Control - Convirtieron: {len(control1)}")
print(f"Control - No convirtieron: {len(control0)}")

tasa_tratamiento = len(tratamiento1) / (len(tratamiento1) + len(tratamiento0))
tasa_control = len(control1) / (len(control1) + len(control0))


print(f"\nTasas de conversión:")
print(f"Tratamiento: {tasa_tratamiento:.3f} ({tasa_tratamiento*100:.1f}%)")
print(f"Control: {tasa_control:.3f} ({tasa_control*100:.1f}%)")
print(f"Diferencia: {tasa_tratamiento - tasa_control:.3f}")
```

```python
tratamiento = chkout[chkout['variante'] == 'tratamiento']['convirtio']
control = chkout[chkout['variante'] == 'control']['convirtio']

ttest_ind(tratamiento, control)

t_stat, p_value = ttest_ind(tratamiento, control)

print(f"Estadístico t: {t_stat:.5f}")
print(f"Valor P: {p_value:.5f}")

if p_value < 0.05:
    print("✅ Diferencia estadísticamente significativa")
else:
    print("❌ No hay diferencia estadísticamente significativa")
```

**INTERPRETACIÓN**

* H0: No hay diferencia real entre la nueva UI y la UI actual".
* Resultado: No hay evidencia estadística suficiente para decir que la nueva UI es mejor.
* Interpretación: La diferencia observada (16.3% vs 15.7%) podría ser simplemente casualidad. 

**RECOMENDACIÓN**

Con un p-value de 0.416, NO se recomienda implementar la nueva UI debido a que no se pueden asegurar mejoras constantes.
